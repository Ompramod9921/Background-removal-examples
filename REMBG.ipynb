```python
# Import necessary libraries
from rembg import remove
from PIL import Image
import matplotlib.pyplot as plt
import io
import os
```

```python
# Directory containing input images
input_directory = r"C:\Users\HP\Desktop\BG remover\test images"

# Get a list of image filenames in the directory
image_filenames = [filename for filename in os.listdir(input_directory) if filename.endswith((".png", ".jpg", ".jpeg"))]
```

```python
# Loop through each image and remove the background
for image_filename in image_filenames:
    # Load the input image using PIL
    input_path = os.path.join(input_directory, image_filename)
    with open(input_path, "rb") as inp_file:
        input_image = Image.open(inp_file).convert("RGBA")

    # Convert the input image to bytes
    input_image_bytes = io.BytesIO()
    input_image.save(input_image_bytes, format="PNG")
    input_image_bytes.seek(0)

    # Remove background using rembg
    output_image_data = remove(input_image_bytes.read())

    # Create a PIL Image from the output image data
    output_image = Image.open(io.BytesIO(output_image_data))

    # Display the input and output images using matplotlib
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(input_image)
    plt.title("Input Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(output_image)
    plt.title("Output Image (Background Removed)")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    print("\n","="*100)
```
